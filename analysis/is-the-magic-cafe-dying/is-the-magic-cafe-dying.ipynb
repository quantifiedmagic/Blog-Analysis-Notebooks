# Is the Magic Caf√© Dying?

```python
import json
import datetime
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = []
with open('../../data/themagiccafe_all_clean.jsonlines') as input_file:
    for row in input_file:
        data = json.loads(row)
        if 'title' not in data:
            continue
        if len(data['title']) < 2:
            continue
        for p in data['posts']:
            
            date = datetime.datetime.strptime(p['date'], 'Posted: %b %d, %Y %H:%M %p')
            dataset.append([date, data['title'][1]])
        
dataset = pd.DataFrame(dataset, columns=['date', 'forum'])
```

```python
dataset['date'] = dataset['date'].astype('datetime64')
dataset['year'] = dataset['date'].dt.year
dataset['month'] = dataset['date'].dt.month
dataset['day'] = dataset['date'].dt.weekday
```

```python
plt.figure(figsize=(10,5))
chart = sns.countplot(x='year', data=dataset, color='#6890F0')
#chart.set_xticklabels(chart.get_xticklabels(), rotation=45)
plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='x-large'  
)
plt.yticks( 
    horizontalalignment='right',
    fontweight='light',
    fontsize='x-large'  
)
plt.xlabel('Year', fontweight='light', fontsize=18)
plt.ylabel('Post Count', fontweight='light', fontsize=18)
plt.annotate( 'Last Special Guest', xy=(15,130000), xytext=(18, 250000), ha='right',  va="center",\
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3,rad=-0.2"))
plt.annotate( 'r/Magic Created', xy=(7,510000), xytext=(12, 500000), ha='right',  va="center",\
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3,rad=-0.2"))
# https://www.theory11.com/forums/
plt.annotate( 'Theory11 Forum Started', xy=(6,500000), xytext=(5, 500000), ha='right',  va="center",\
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3,rad=-0.2"))
plt.title('Magic Caf√© Posts across Time', fontsize=24)
plt.savefig('magic-cafe-time.png')
```
